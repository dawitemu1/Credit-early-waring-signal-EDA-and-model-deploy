Boxplots are useful, as they provide a visual summary of the data enabling researchers to quickly identify mean values, the dispersion of the data set, and signs of skewness.

 **Interpretation of  a boxplot**. 
 
A boxplot always contains a box with `two lines above and below it`. Those lines are called **whiskers**, and the ends of those lines represent the upper and lower bounds of the dataset. There are also some values above the top whisker,which are outliers. Notice that the plot divides the data into four sections: 

- the bottom of the lower whisker to the bottom of the box, 
- the bottom of the box to the line in the middle, 
- the middle line to the top line of the box, and 
- the top of the box to the end of the upper whisker. 

These sections are called **intervals**, and the three lines that divide them are called **quartiles**. Each interval contains 25% of the observations in the dataset, which means that the box created by the first and third quartiles represents the middle 50% of observations. The second quartile — the orange line on our graph — represents the dataset's median value. 

Keeping all that in mind, the boxplot here shows that while there are some very large properties in our Colombia dataset, half the properties tend to be much smaller. Remember those data points above the top whisker? Those are called **outliers**, and they represent values so extreme that they fall outside the meaningful spread of observations in the dataset. 

- Part of the purpose of making a boxplot is to find those outliers and resolve them from future analyses.

## Histograms

A histogram is a graph that shows the frequency distribution of numerical data. In addition to helping us understand frequency, histograms are also useful for detecting outliers.

Let's create a histogram using the `sessame_close_df`  dataframe.

```python
plt.hist(sessame_close_df, bins=10, rwidth=0.9, color="b")
plt.title("Closiing Price of Sessame")
plt.xlabel("Close")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.75);
```

There are two things to consider in this histogram. First and foremost is its interpretation: 
1. the distribution is skewed significantly to the right. In this case, the data suggest that the majority of properties is to the right

2. You might also have noticed that there are ten bars. In a histogram, we call these bars **bins**. A bin is simply a way to group data to make it easier to see trends. You can use as many or as few as you like; just recognize that the fewer bins you use, the less detailed the output will become. 
             
Let's take a look at what the same data looks like when it's in 20 bins instead of 10.

```python
plt.hist(sessame_close_df, bins=20, rwidth=0.9, color="b")
plt.title("Closiing Price of Sessame")
plt.xlabel("Close")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.75);
```

### Histogram, Distribution Graph and Box plot on subplot

Gnerate a function that can acept the dataset,and name of feature so that it can plot Distribution Graph, Histogram and Box plot

```python
import matplotlib.pyplot as plt

# Function to Plot the chart for numerical independent feature
def plotGraph(dataset, feature):
    plt.style.use('fivethirtyeight')
    # Create a new dataset without missing values
    filtered_data = dataset.dropna(subset=[feature])
    filtered_data[feature] = filtered_data[feature].astype(float)  # Convert to float

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.title(f'{feature} Distribution Graph', size=12)
    plt.hist(filtered_data[feature], color="red", density=True)

    plt.subplot(1, 3, 2)
    plt.title(f'{feature} Histogram Graph', color="red", size=12)
    plt.hist(filtered_data[feature], color='red', density=True, bins=10)
    plt.xlabel(feature)

    plt.subplot(1, 3, 3)
    plt.boxplot(dataset[feature], vert=True, widths=0.7, patch_artist=True)
    plt.title(f'Box Plot of {feature}', size=12)
    plt.xlabel(feature)
    plt.subplots_adjust(wspace=0.3)
    plt.show()
```

```python
# Load the sessame  dataset
sessame_df = pd.read_csv("./Dataset/sessame.csv")
```

```python
# call the plotGraph  for sessame_df  adatset and Close Feature
plotGraph(sessame_df,"Close")
```

```python
# call the plotGraph  for sessame_df  adatset and High Feature
plotGraph(sessame_df,"High")
```

## Scatter Plots

A scatter plot is a graph that uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables, and are especially useful if you're looking for correlations.

Let's create a simple scatter plot using the "sessame" data .

```python

plt.scatter(sessame_df["Vol_in_Ton"], sessame_df["Close"], color="r")
plt.xlabel("Volume in Tones")
plt.ylabel("Closing price")
plt.title("Volume in Tones Vs Closing price");
```
